## Bivariate Data

### a. Line Plot

```python
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,36,49]
z = [1,8,27,64,125,216,343]
```

```python
plt.figure(figsize = (5,5)) ## To control the size of figure
plt.title('Line Plot',fontsize =15) ## Specifies Title of Plot
plt.xlabel('X-Values') ## Specifes the X-labels
plt.ylabel('Y-Values') ## Specifes the Y-labels

plt.plot(x,y,color = 'blue',
         linestyle = 'dashdot',
        marker = '^',
        markersize = 12,
        markerfacecolor = '#b3be54') ## This is ploting the data
plt.show()    ## This plots as per matplotlib's way
```

### Plotting more than one Curve in the Same figure

```python
### Some Hypothetical plot
plt.figure(figsize = (5,5))
plt.title('Line Plot',fontsize =15)
plt.xlabel('Decades')
plt.ylabel('Social Maturity')

## Plot for Cube of Numbers
plt.plot(x,z,color = 'orange',
         linestyle = 'solid',
        marker = '^',
        markersize = 10,
        markerfacecolor = '#581fd2',
        label = 'India')

## Plot for Square of Numbers
plt.plot(x,y,color = 'blue',
         linestyle = 'dashdot',
        marker = 'H',
        markersize = 10,
        markerfacecolor = '#b3be54',
        label = 'Pakistan')


plt.legend()
plt.show()    
```

### b.Scatter Plot

```python
price = pd.read_csv('datasets/scatter_data.csv')
price.head()
```

#### Rename Columns within DataFrame

```python
price.rename(columns = {'Area (ft.)':'Area','Year of sale':'SaleYear'},inplace = True)
price.head()
```

```python
plt.figure(figsize = (5,5))
plt.title('House Prices')
plt.xlabel('Area In Sq. Ft')
plt.ylabel('Price')
plt.scatter(price['Area'],price['Price'],
           color = '#00ff00',
            edgecolor = '#ffebcd',
           marker = 'H',
           s=20)
plt.show()
```

```python
CarsListings = pd.read_csv('datasets/bar_chart_data.csv')
CarsListings.head()
```

```python
CarsListings.rename(columns = {'Cars Listings' : 'Listings'},inplace = True)
CarsListings.head()
```

## Numerical V/s Categorical Data
### a.BarChart 

```python
plt.figure(figsize = (10,5))
plt.title("Cars Listed For Each Brand", fontdict = {
    'fontsize' : 14,
    'color': '#b0c24a'})
plt.bar(CarsListings['Brand'],
        CarsListings['Listings'],
       color = '#ccff00') # the bar() function is plotting the bar chart
plt.xticks(fontsize = 10 
           ,rotation = 45
          )
plt.show()
```

## b.The horizontal Bar

```python
plt.figure(figsize = (10,5))
plt.title("Cars Listed For Each Brand")
plt.barh(CarsListings['Brand'],
        CarsListings['Listings'],
       color = '#ee82ee')
plt.yticks(fontsize = 14
          )
plt.xticks(fontsize = 14 
           ,rotation = 45
          )
plt.show()
```

## Pie Plot 

```python
plt.pie(CarsListings['Listings'],
       labels = CarsListings['Brand'],
        autopct = '%0.1f%%',
        textprops={'fontsize':10,'color':'#000039'},
        explode = (0,0,0,0,0,0,0.1)
       )
plt.show()
```

## Cat plot
1. Bar Plot

```python
df_cars = pd.read_csv('datasets/bar_chart_data.csv')
df_cars
```

## Seaborn Works Hand in Hand with matplotlib

```python
# pal = sns.color_palette('viridis')
sns.catplot(df_cars,x = 'Brand',y = 'Cars Listings',
            kind = 'bar',
            aspect =2/1,
           palette = sns.color_palette('viridis') # Can pass pal also 
           ) ## Plots
plt.show() # Renders the Plot
```

## Horizontal bar is not supported in seaborn but can be obtained interchanging x and y in bar plot 

```python
# pal = sns.color_palette('viridis')
sns.catplot(df_cars,x = 'Cars Listings',y = 'Brand',
            kind = 'bar',
            # aspect =2/1,
           palette = sns.color_palette('viridis') # Can pass pal also 
           ) ## Plots
plt.show() # Renders the Plot
```

### Box Plot

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill',
           kind = 'box')
plt.show()
```

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill',
           kind = 'box',
           hue = 'time')
plt.show()
```

## Strip plot

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill',
           kind = 'strip',
           # hue = 'time',
        palette = 'Set2'
           )
plt.show()
```

## Swarm Plot

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill',
           kind = 'swarm', 
           # hue = 'time',
           palette = 'Set2'
           )
plt.show()
```

### Relplot

```python
mpg = pd.read_csv('datasets/mpg.csv')
mpg
```

### line plot

```python
sns.relplot(data=mpg, x = 'model_year', y = 'mpg',
           kind = 'line',
           ci = False,
           palette = 'Set2',
           hue = 'origin',
           style = 'origin')
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill',bins=10,
            element = 'step',fill = False,
            color = 'black')
plt.show()
```
