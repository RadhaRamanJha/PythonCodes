## Bivariate data -- Numeric Verses vs Numeric data

+ a. Line Plot

```python
x = [1,2,3,4,5,6,7]
y = [2,4,9,16,25,36,49]
```

```python
plt.plot(x,y,
        linestyle = 'dotted',
        color = 'midnightblue',
        marker = '^',
        markersize =8,
        markerfacecolor = 'magenta')
plt.title('Line Plot',fontsize= 18, color = 'blue')
plt.xlabel('X-values', color = 'green')
plt.ylabel('Y-values',color = 'brown')
plt.show()
```

## Scatter Plot

```python
prices = pd.read_csv('datasets/house_prices.csv')
prices.head()
```

```python
prices.rename(columns={'Area (ft.)':'Area'},inplace = True)
```

```python
plt.scatter(prices['Area'],
            prices['Price'],
           color = 'yellow',
           edgecolor = 'midnightblue')
plt.title('House Prices',fontsize = 14)
plt.xlabel('Area in Sq.Ft')
plt.ylabel('HousePrice')
plt.show()
```

## Numeric Vs Categorical

```python
courses = ['Data Science','Data Analysis','My SQL', 'Tableau', 'Power Bi']
students = [89,103,76,34,92]
```

```python
plt.figure(figsize = (10,6))
plt.bar(x = courses, height = students,
       color = '#ffa07a',
       edgecolor = 'blue',
       linewidth = 2)
plt.title('Course Details',fontsize = 14)
plt.xlabel('Courses')
plt.ylabel('Students')
plt.xticks(fontsize = 8,
          rotation = 45)
plt.show()
```

## Horizontal Bar

```python
plt.figure(figsize = (6,3))
plt.barh(y = courses, width = students,
       color = '#ffa07a',
       edgecolor = 'blue',
       linewidth = 2)
plt.title('Course Details',fontsize = 14)
plt.xlabel('Students')
plt.ylabel('Courses')
plt.xticks(fontsize = 8,
          rotation = 45)
plt.show()
```

## Pie Chart

```python
courses = ['Data Science','Data Analysis','My SQL', 'Tableau', 'Power Bi']
students = [89,103,76,34,92]
```

```python
plt.pie(students, labels = courses,
       autopct='%0.1f%%',
       textprops={'fontsize':8,'color':'black' },
       wedgeprops={'linewidth':1,'edgecolor':'black'},
       explode = [0.1,0,0,0,0],
       colors = ['#c32148','#008b8b','#0000ff','#d8bfd8','#db7093'])
plt.title('Student Course Choices',fontsize = 14)
plt.show()
```

## Histogram 

```python
plt.hist(prices['Price'],edgecolor = 'white',
        color = '#7fffd4',
        alpha = 0.5)
plt.title('Price Details',fontsize = 14)
plt.xlabel('Price Range')
plt.ylabel('Number of Houses')
plt.show()
```
